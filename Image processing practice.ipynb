```python
def im_hist(img):
    fig = plt.figure(figsize=(88,16))
    fig.clf()
    ax=fig.gca()
    ax.hist(img.flatten(), bins = 256)
    plt.show()

im_hist(i_unicolor) #visualizing pixel value distributions in the image
```

```python
from skimage import exposure #equalizing the image uning sklearn

i_eq = exposure.equalize_hist(i_unicolor)
imshow(i_eq, cmap = 'gray')
```

```python
import skimage #adding noise to the image
i_n = skimage.util.random_noise(i_eq)
imshow(i_n, cmap='gray')
```

```python
def gauss_filter(im, sigma=10): #using a Guassian filter on the image
    from scipy.ndimage.filters import gaussian_filter as gf
    import numpy as np
    return gf(im, sigma=sigma)
i_g = gauss_filter(i_n)
imshow(i_g, cmap = 'gray')
```

```python
def img_hist(img):
    fig = plt.figure(figsize=(16,6))
    fig.clf()
    ax = fig.gca()
    ax.hist(img.flatten(), bins = 256)
    plt.show()
    
img_hist(j_unicolor)
```

```python
#visualizing the cumulative histogram of pixel distribution 
```

```python
def cum_hist(img):
    fig = plt.figure(figsize=(16,6))
    fig.clf()
    bx = fig.gca()
    bx.hist(img.flatten(), bins = 256, cumulative = 'True')
    plt.show()
    
cum_hist(j_unicolor)
```
